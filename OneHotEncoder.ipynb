```python
import pandas as pd
import numpy as np

data = {
    'City': ['New York', 'Aditya', 'New York', 'Chicago', 'Aditya', 'Chicago'],
    'Temperature': ['Hot', 'Cold', 'Warm', 'Cold', 'Warm', 'Hot']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
def ordinal_encode(df, column, order):
    for i in range(len(df[column])):
            for rank, value in enumerate(order):
                if df[column][i] == value:
                    df.at[i, column] = rank
                    break
    return df
```

```python
temperature_order = ['Cold', 'Warm', 'Hot']
df_ord_encoded = ordinal_encode(df, 'Temperature', temperature_order)
df_ord_encoded
```

```python
def one_hot_encode(df, column): 
    unique_vals = df[column].unique()
    for val in unique_vals:
        df[f"{column}_{val}"] = np.where(df[column] == val, 1, 0)
    df = df.drop(column, axis=1)
    return df
```

```python
df_one_encoded = one_hot_encode(df, 'City')
```

```python
df_one_encoded
```

```python
df = df.drop(columns=[col for col in df.columns if 'City_' in col], errors='ignore')
one_hot_encoded_data = pd.get_dummies(df, columns=['City'])
one_hot_encoded_data = one_hot_encoded_data.astype(int)
```
